```python
#Importing Libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('drugs.csv')
df_new = df[['Rates.Tobacco.Use Past Month.18-25','Rates.Illicit Drugs.Cocaine Used Past Year.18-25']]
df_final = df_new.rename(columns={'Rates.Tobacco.Use Past Month.18-25': 'Tobacco Use', 'Rates.Illicit Drugs.Cocaine Used Past Year.18-25': 'Cocaine Use'})
df_final
```

```python
df_final.plot.scatter('Cocaine Use', 'Tobacco Use')
```

```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
df_final.corr(method= histogram_intersection)
```

```python
df_alc_coke = df[['Totals.Alcohol.Use Disorder Past Year.26+','Totals.Illicit Drugs.Cocaine Used Past Year.26+']]
df_final2 = df_alc_coke.rename(columns={'Totals.Alcohol.Use Disorder Past Year.26+': 'Alcohol Disorder, Age 26+', 'Totals.Illicit Drugs.Cocaine Used Past Year.26+': 'Cocaine Use, Age 26+'})
df_final2
```

```python
df_final2.plot.scatter('Alcohol Disorder, Age 26+', 'Cocaine Use, Age 26+')
```

```python
r = np.corrcoef(df_final2['Alcohol Disorder, Age 26+'], df_final2['Cocaine Use, Age 26+'])
print(r)
```
